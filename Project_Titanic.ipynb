```python
import pandas as pd
import numpy as np
```

```python
df1=pd.read_csv('gender_submission.csv')
df1
```

```python
df2=pd.read_csv('test.csv')
df2
```

```python
df3=pd.read_csv('train.csv')
df3
```

```python
#объединить данные df1 и df2
df_1_2=pd.merge(df1, df2, on='PassengerId', how='left')
df_1_2
```

```python
# конкатинация df_1_2 и df3
df=pd.concat([df3,df_1_2], ignore_index=True)
df
```

```python
df.info()
```

-  Очистка данных

```python
df.isnull().sum()
```

```python
#- Столбец Age: Заполните пропуски  средним значением в зависимости от пассажирского класса и пола.
df['Age']=df['Age'].fillna(df.groupby(['Pclass','Sex'])['Age'].transform('mean'))
df.isnull().sum()
```

```python
df['Age'] = df['Age'].round().astype(int)
df
```

```python
df['Age'].unique()
```

```python
# Заполняем пропуски в 'Embarked' наиболее частым значением
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df['Embarked'].isnull().sum() 
```

```python
#- Столбец Cabin: удаляем столбец полностью.
df=df.drop('Cabin', axis=1)
df
```

```python
#-Столбец Fare:Заполяем их медианным значением.
df['Fare']=df['Fare'].fillna(df['Fare'].median())
df.isnull().sum()
```

- Создание новых признаков. Feature Engineering

```python
# Извлекаем титул из столбца Name и создаем новый столбец Title
df['Title']=df['Name'].str.extract(r',\s*([^\.]+)\.',expand=False)
df
```

```python
# Создаем новый столбец FamilySize как сумму SibSp, Parch и 1 (самого пассажира)
df['FamilySize']=df['SibSp'] + df['Parch'] + 1
df
```

```python
# Создаем бинарный признак IsAlone: 1, если пассажир один, иначе 0
df['IsAlone']=(df['FamilySize']==1).astype(int)
df
```

```python
df.to_csv('your_table.csv', index=False)
```

6. Круговая диаграмма для пункта отправления (Embarked)

```python
survived_counts = df[df['Survived'] == 1]['Embarked'].value_counts()                    #Сумма выживших поссажиров
not_survived_counts = df[df['Survived'] == 0]['Embarked'].value_counts()                #Сумма не выживших поссажиров
fig, ax = plt.subplots(1, 2, figsize=(15, 7))                                           #Размещаем диаграммы 
survived_counts.plot(kind='pie', autopct='%1.1f%%', ax=ax[0], title="Процент выживших по пунктам отправления")          #Процент выживших по пунктам отправления
not_survived_counts.plot(kind='pie', autopct='%1.1f%%', ax=ax[1], title="Процент не выживших по пунктам отправления")   #Процент не выживших по пунктам отправления
plt.show()
```

8. Тепловая карта корреляций (heatmap)

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])                 #выбирает только числовые столбцы
correlation_matrix = numeric_df.corr()                                      #вычисляет корреляционную матрицу(-1:1)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', square=True)   # sns.heatmap - создает тепловую карту; 
plt.title("Корреляционная матрица признаков")                                    # annot=True - отображает значения корреляций на ячейках тепловой карты
plt.show()                                                                                #   cmap='coolwarm' -  задает цветовую схему от холодных к теплым 
                                                                                                   #  square=True - делает ячейки квадратными, придавая карте аккуратный вид
```

- Сортировка и группировка данных. Data Aggregation & Grouping

```python
#Выполните группировку по таким признакам, как Pclass, Sex, Embarked и FamilySize, чтобы изучить средний показатель выживаемости для каждой группы.
avg_survival = df.groupby(['Pclass','Sex','FamilySize'])['Survived'].mean()
avg_survival_df = avg_survival.reset_index().rename(columns={'Survived':'Average Survival Rate'})
avg_survival_df.head()
```

```python
#Используйте группировку и сортировку для анализа зависимости между стоимостью билета и классом (Pclass)
fare_by_class = df.groupby('Pclass')['Fare'].mean().sort_values(ascending=False)
fare_by_class
```

```python
#Используйте группировку и сортировку для анализа зависимости между размером семьи и выживаемостью
survival_by_family = df.groupby('FamilySize')['Survived'].mean().sort_values(ascending=False)
survival_by_family
```
